```python
A=df.sort_values(by = "Active",ascending =False)
```

```python
A.head(10)
```

```python
ax=sns.countplot(x='Active',data=A)
for bars in ax.containers:
    ax.bar_label(bars)
    ax.set_xlim(0,10)
```
